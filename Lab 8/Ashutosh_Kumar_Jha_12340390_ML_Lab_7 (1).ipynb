```python
# =======================
# Imports
# =======================
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# For reproducibility
np.random.seed(42)
```

```python
# =======================
# 1. Define the Function to Optimize
# =======================
def func(x, y):
    return 0.5 * x**2 + 5 * y**2

def grad(x, y):
    return np.array([x, 10 * y])
```

```python
# =======================
# 2. Optimizer Implementations
# =======================
```

```python
class SGD:
    def __init__(self, lr=0.05):
        self.lr = lr

    def step(self, params, grads):
        return params - self.lr * grads
```

```python
class SGDMomentum:
    def __init__(self, lr=0.05, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = np.zeros_like(grads := np.zeros(2))

    def step(self, params, grads):
        self.v = self.momentum * self.v - self.lr * grads
        return params + self.v
```

```python
class NAG:
    def __init__(self, lr=0.1, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = np.zeros(2)

    def step(self, params, grads_fn):
        look_ahead = params + self.momentum * self.v
        grads = grads_fn(*look_ahead)
        self.v = self.momentum * self.v - self.lr * grads
        return params + self.v
```

```python
class Adagrad:
    def __init__(self, lr=0.3, eps=1e-8):
        self.lr = lr
        self.eps = eps
        self.cache = np.zeros(2)

    def step(self, params, grads):
        self.cache += grads ** 2
        return params - self.lr * grads / (np.sqrt(self.cache) + self.eps)
```

```python
class RMSProp:
    def __init__(self, lr=0.05, beta=0.9, eps=1e-8):
        self.lr = lr
        self.beta = beta
        self.eps = eps
        self.cache = np.zeros(2)

    def step(self, params, grads):
        self.cache = self.beta * self.cache + (1 - self.beta) * grads ** 2
        return params - self.lr * grads / (np.sqrt(self.cache) + self.eps)
```

```python
class Adam:
    def __init__(self, lr=0.05, beta1=0.9, beta2=0.999, eps=1e-8):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.eps = eps
        self.m = np.zeros(2)
        self.v = np.zeros(2)
        self.t = 0

    def step(self, params, grads):
        self.t += 1
        self.m = self.beta1 * self.m + (1 - self.beta1) * grads
        self.v = self.beta2 * self.v + (1 - self.beta2) * (grads ** 2)

        m_hat = self.m / (1 - self.beta1 ** self.t)
        v_hat = self.v / (1 - self.beta2 ** self.t)

        return params - self.lr * m_hat / (np.sqrt(v_hat) + self.eps)
```

```python
def optimize(optimizer, name, steps=50, start=np.array([2.0, 2.0])):
    params = start.copy()
    trajectory = [params.copy()]
    for i in range(steps):
        if isinstance(optimizer, NAG):
            params = optimizer.step(params, grad)
        else:
            g = grad(*params)
            params = optimizer.step(params, g)
        trajectory.append(params.copy())
    trajectory = np.array(trajectory)

    # Plot
    x = np.linspace(-3, 3, 100)
    y = np.linspace(-3, 3, 100)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    plt.figure(figsize=(5, 4))
    plt.contour(X, Y, Z, levels=30, cmap=cm.viridis)
    plt.plot(trajectory[:, 0], trajectory[:, 1], 'r.-', label='Path')
    plt.title(f"{name} Optimization Path")
    plt.xlabel('x'); plt.ylabel('y')
    plt.legend()
    plt.show()
```

```python
print("Visualizing different optimizers on an anisotropic function...\n")

optimize(SGD(lr=0.05), "SGD")
optimize(SGDMomentum(lr=0.05), "SGD + Momentum")
optimize(NAG(lr=0.05), "Nesterov Accelerated Gradient")
optimize(Adagrad(lr=0.3), "Adagrad")
optimize(RMSProp(lr=0.05), "RMSProp")
optimize(Adam(lr=0.05), "Adam")
```

```python
def loss_plot(optimizer, name, steps=50, start=np.array([2.0, 2.0]), plot_path=True):
    """
    Runs optimization using the given optimizer.
    Returns the loss trajectory and optionally plots optimization path.
    """
    params = start.copy()
    trajectory = [params.copy()]
    losses = [func(*params)]

    for i in range(steps):
        if isinstance(optimizer, NAG):
            params = optimizer.step(params, grad)
        else:
            g = grad(*params)
            params = optimizer.step(params, g)

        trajectory.append(params.copy())
        losses.append(func(*params))

    trajectory = np.array(trajectory)

    # Plot optimization path (contour)
    if plot_path:
        x = np.linspace(-3, 3, 100)
        y = np.linspace(-3, 3, 100)
        X, Y = np.meshgrid(x, y)
        Z = func(X, Y)

        plt.figure(figsize=(5, 4))
        plt.contour(X, Y, Z, levels=30, cmap=cm.viridis)
        plt.plot(trajectory[:, 0], trajectory[:, 1], 'r.-', label='Path')
        plt.title(f"{name} Optimization Path")
        plt.xlabel('x'); plt.ylabel('y')
        plt.legend()
        plt.show()

    return np.array(losses)
```

```python
# Compare convergence rates
optimizers = {
    "SGD": SGD(lr=0.05),
    "Momentum": SGDMomentum(lr=0.05),
    "NAG": NAG(lr=0.05),
    "Adagrad": Adagrad(lr=0.3),
    "RMSProp": RMSProp(lr=0.05),
    "Adam": Adam(lr=0.05),
}

loss_histories = {}

# Disable contour plot for faster run when comparing
for name, opt in optimizers.items():
    print(f"Running {name}...")
    loss_histories[name] = loss_plot(opt, name, steps=60, plot_path=False)

# Plot loss convergence
plt.figure(figsize=(8, 6))
for name, losses in loss_histories.items():
    plt.plot(losses, label=name)
plt.yscale('log')
plt.title("Loss Convergence Comparison")
plt.xlabel("Iteration")
plt.ylabel("Loss (log scale)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# =======================
# 5. Initialization Demonstration
# =======================
def visualize_initializations():
    n_neurons = 256
    fan_in = 256
    fan_out = 128

    # Different initialization schemes
    initializations = {
        "Random Normal": np.random.randn(n_neurons),
        "Glorot (Xavier)": np.random.randn(n_neurons) * np.sqrt(2.0 / (fan_in + fan_out)),
        "He": np.random.randn(n_neurons) * np.sqrt(2.0 / fan_in),
        "LeCun": np.random.randn(n_neurons) * np.sqrt(1.0 / fan_in)
    }

    plt.figure(figsize=(10, 6))
    for name, vals in initializations.items():
        plt.hist(vals, bins=50, alpha=0.6, label=name, density=True)
    plt.title("Weight Initialization Distributions")
    plt.legend()
    plt.show()


visualize_initializations()
```

```python
# ===============================
# EXERCISE 1: LR & Momentum Tests
# ===============================

learning_rates = [0.001, 0.01, 0.1]
momentums = [0.0, 0.5, 0.9]

def run_experiment(optimizer, name):
    losses = loss_plot(optimizer, name, steps=50, plot_path=False)
    plt.plot(losses, label=name)

plt.figure(figsize=(10,6))
for lr in learning_rates:
    run_experiment(SGD(lr=lr), f"SGD (lr={lr})")

for lr in [0.01, 0.05]:
    for m in momentums:
        run_experiment(SGDMomentum(lr=lr, momentum=m), f"Momentum (lr={lr}, m={m})")

for lr in [0.01]:
    for m in [0.9]:
        run_experiment(NAG(lr=lr, momentum=m), f"NAG (lr={lr}, m={m})")

plt.title("Effect of LR & Momentum on Convergence")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.yscale("log")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ===============================
# EXERCISE 2: Implement AdamW
# ===============================

class AdamW:
    def __init__(self, lr=0.05, beta1=0.9, beta2=0.999, weight_decay=0.01, eps=1e-8):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.weight_decay = weight_decay
        self.eps = eps
        self.m = np.zeros(2)
        self.v = np.zeros(2)
        self.t = 0

    def step(self, params, grads):
        self.t += 1

        # Decoupled weight decay
        params = params - self.lr * self.weight_decay * params

        self.m = self.beta1 * self.m + (1 - self.beta1) * grads
        self.v = self.beta2 * self.v + (1 - self.beta2) * (grads ** 2)

        m_hat = self.m / (1 - self.beta1 ** self.t)
        v_hat = self.v / (1 - self.beta2 ** self.t)

        return params - self.lr * m_hat / (np.sqrt(v_hat) + self.eps)

# ---- Visualize AdamW ----
adamw_losses = loss_plot(AdamW(lr=0.05, weight_decay=0.01), "AdamW", steps=60, plot_path=True)
```

```python
# ===============================
# EXERCISE 3: Initialization Tests
# ===============================

def test_initialization(init_fn, name):
    W = np.random.randn(256, 256)
    W = init_fn(W)

    x = np.random.randn(1, 256)         # input
    out = x @ W                         # forward pass
    variance = np.var(out)

    print(f"{name}: output variance = {variance:.5f}")
    return out

initializations = {
    "Random Normal": lambda W: np.random.randn(*W.shape),
    "Xavier": lambda W: np.random.randn(*W.shape) * np.sqrt(2/(256+256)),
    "He": lambda W: np.random.randn(*W.shape) * np.sqrt(2/256),
    "LeCun": lambda W: np.random.randn(*W.shape) * np.sqrt(1/256),
}

results = {}
for name, fn in initializations.items():
    results[name] = test_initialization(fn, name)

# ---- Plot distributions ----
plt.figure(figsize=(10,6))
for name, out in results.items():
    plt.hist(out.flatten(), bins=40, alpha=0.5, label=name, density=True)

plt.title("Output Distribution After One Forward Pass")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# ===============================
# BONUS: Different loss landscapes
# ===============================

def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

def rosenbrock_grad(x, y):
    dx = -2*(1-x) - 400*x*(y - x**2)
    dy = 200*(y - x**2)
    return np.array([dx, dy])

def optimize_func(f, g, optimizer, name, steps=60, start=np.array([-1.5, 1.5])):
    params = start.copy()
    path = [params.copy()]

    for _ in range(steps):
        grads = g(*params)
        params = optimizer.step(params, grads)
        path.append(params.copy())

    path = np.array(path)

    # Contour plot
    x = np.linspace(-2, 2, 200)
    y = np.linspace(-1, 3, 200)
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)

    plt.figure(figsize=(6,5))
    plt.contour(X, Y, Z, levels=50, cmap='viridis')
    plt.plot(path[:,0], path[:,1], 'r.-')
    plt.title(f"{name} on Rosenbrock Function")
    plt.show()

optimize_func(rosenbrock, rosenbrock_grad, SGD(lr=0.001), "SGD")
optimize_func(rosenbrock, rosenbrock_grad, Adam(lr=0.01), "Adam")
```
