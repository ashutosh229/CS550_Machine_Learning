```python
for breed in os.listdir(train_dir):
    breed_train_path = os.path.join(train_dir, breed)
    if not os.path.isdir(breed_train_path):
        continue
    breed_val_path = os.path.join(val_dir, breed)
    os.makedirs(breed_val_path, exist_ok=True)
    images = os.listdir(breed_train_path)
    random.shuffle(images)
    num_val = int(len(images) * val_split)
    val_images = images[:num_val]
    print(f"Moving {num_val} images for breed {breed} to validation set.")
    for img in val_images:
        src = os.path.join(breed_train_path, img)
        dest = os.path.join(breed_val_path, img)
        shutil.move(src, dest)
```

## Part 1 â€” Data & Model Setup

### Requirements
1. **Data Loading & Augmentation**
   - Assume a dataset is organized using `ImageFolder` structure (e.g., `data/train/black_dog/`, `data/train/red_brown_dog/`, etc.).
   - Implement **data augmentation** for the training set (e.g., `RandomResizedCrop`, `RandomHorizontalFlip`).
   - Implement **standard transforms** for the validation/test set (e.g., `Resize`, `CenterCrop`).
   - Apply `Normalize` transform (using ImageNet mean and std) to all sets.

2. **Model Architecture**
   - Load a **ResNet-18** model pre-trained on ImageNet.
   - **Freeze** the parameters of all layers *except* the final fully-connected layer.
   - **Replace** the final fully-connected layer (`fc`) with a new one that matches the number of classes in your custom dataset.


```python
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std  = [0.229, 0.224, 0.225]
```

```python
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std)
])
```

```python
test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=imagenet_mean, std=imagenet_std)
])
```

```python
train_dir = "/content/dataset/Dog-Breed-Classification/train"
val_dir   = "/content/dataset/Dog-Breed-Classification/val"
test_dir  = "/content/dataset/Dog-Breed-Classification/test"
```

```python
train_data = datasets.ImageFolder(train_dir, transform=train_transforms)
val_data   = datasets.ImageFolder(val_dir, transform=test_transforms)
test_data  = datasets.ImageFolder(test_dir, transform=test_transforms)
```

```python
train_loader = DataLoader(train_data, batch_size=32, shuffle=True, num_workers=2)
val_loader   = DataLoader(val_data, batch_size=32, shuffle=False, num_workers=2)
test_loader  = DataLoader(test_data, batch_size=32, shuffle=False, num_workers=2)
```

```python
model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
```

```python
for param in model.parameters():
    param.requires_grad = False
```

```python
num_classes = len(train_data.classes)
```

```python
model.fc = nn.Linear(model.fc.in_features, num_classes)
```

```python
for name, param in model.named_parameters():
    print(name, param.requires_grad)
```

## Part 2 â€” Training Loop

### Requirements
Implement a full **training loop** that includes:

- **Loss:** Use **Cross-Entropy Loss** (`nn.CrossEntropyLoss`).
- **Optimizer:** Implement the **Adam optimizer** (`torch.optim.Adam`).
   - *Crucially*, ensure the optimizer is *only* passed the parameters of the *new final layer* that you created.
- **Validation:** After each training epoch, run a full validation loop to check performance.
- **Model Saving:** Save the best model (the `state_dict`) based on the **highest validation accuracy**.
- **Logging:** Print Training Loss, Training Accuracy, Validation Loss, and Validation Accuracy for each epoch.

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
optimizer = optim.Adam(model.fc.parameters(), lr=1e-3)
```

```python
num_epochs = 10
best_val_acc = 0.0
save_path = "best_resnet18.pth"
```

```python
def calculate_accuracy(logits, labels):
    _, preds = torch.max(logits, 1)
    return torch.sum(preds == labels).item() / len(labels)
```

```python
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    running_acc = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        running_acc  += calculate_accuracy(outputs, labels) * images.size(0)

    epoch_train_loss = running_loss / len(train_loader.dataset)
    epoch_train_acc  = running_acc  / len(train_loader.dataset)

    model.eval()
    val_running_loss = 0.0
    val_running_acc = 0.0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            val_running_loss += loss.item() * images.size(0)
            val_running_acc  += calculate_accuracy(outputs, labels) * images.size(0)

    epoch_val_loss = val_running_loss / len(val_loader.dataset)
    epoch_val_acc  = val_running_acc  / len(val_loader.dataset)

    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        torch.save(model.state_dict(), save_path)

    print(f"Epoch [{epoch+1}/{num_epochs}]")
    print(f"  Train Loss: {epoch_train_loss:.4f} | Train Acc: {epoch_train_acc*100:.2f}%")
    print(f"  Val Loss:   {epoch_val_loss:.4f} | Val Acc:   {epoch_val_acc*100:.2f}%")
    print("-" * 50)
```

```python
classes = test_data.classes
```

```python
def imshow(img_tensor):
    img = img_tensor.cpu().numpy().transpose((1, 2, 0))
    img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    plt.axis("off")
```

```python
examples_shown = 0
```

```python
plt.figure(figsize=(12, 18))

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)

        for i in range(images.size(0)):
            if examples_shown >= 10:
                break

            plt.subplot(5, 2, examples_shown + 1)
            imshow(images[i])

            prediction = classes[preds[i].item()]
            reference  = classes[labels[i].item()]
            result = "Correct" if prediction == reference else "Incorrect"

            plt.title(f"Prediction: {prediction}\nReference: {reference}\nResult: {result}",
                      fontsize=10)

            examples_shown += 1

        if examples_shown >= 10:
            break

plt.tight_layout()
plt.show()
```
